```python
import os 
import sqlalchemy as db
```

```python
db_user = os.environ.get('JOBHUNTER_DB_USER', None)
db_pass = os.environ.get('JOBHUNTER_DB_PASS', None)
```

```python
engine = db.create_engine(f'postgresql+psycopg2://{db_user}:{db_pass}@localhost/jobhunter', echo = True)
connection = engine.connect()
metadata = db.MetaData()
jobs = db.Table('jobs', metadata, autoload = True, autoload_with = engine)
```
